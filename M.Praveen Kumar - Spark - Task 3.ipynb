# Data Cleaning

After see the outliers of the numerical variable we decide to clean the data.
* Removing of outliers
* handle missing values
* remove duplicate values

We remove this outliers by 2 method 
* Z score
* IQR (Inter Quantile range)

# Based on IQR method

```python

Q1 = df_num.quantile(0.25)
Q3 = df_num.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
#Before IQR method
df.shape
```

```python
#IQR method
df_sales = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
#After IQR method
df_sales.shape
```

# Check for outlier after IQR Method

```python
variables = ['Sales','Quantity','Discount', 'Profit']
fig, ax = plt.subplots(1, 4, figsize=(15, 4))
for variable, subplot in zip(df_sales[variables], ax.flatten()):
    z = sns.boxplot(x=df_sales[variable], orient = "h",whis=1.5 , ax=subplot)
    z.set_xlabel(variable, fontsize = 20)
```

Since city and Postal Code is not important for analysis part so i decide to drop this two features for analysis part

```python
retail = df.drop(['City', 'Postal Code'], axis = 1)
```

```python
retail.head()
```

```python
retail.shape
```

```python
#Crt_sales indicate the actual sales - Discount
def crt_sales(retail):
    crt_sales = retail['Sales'] * (1-retail['Discount'])
    return crt_sales
retail['crt_sales'] = retail.apply(crt_sales, axis = 1)
```

```python
retail.head()
```

```python
#Cost is current_sale - Profit
def cost(retail):
    cost = retail['crt_sales'] - (retail['Profit'])
    return cost
retail['cost'] = retail.apply(cost, axis=1)
```

```python
retail.head()
```
